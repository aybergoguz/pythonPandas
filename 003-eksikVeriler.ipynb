## group by

```python
maasSozlugu={"Departman":["Yazılım","Yazılım","Pazarlama","Pazarlama","Hukuk","Hukuk"],
             "Calısan Ismi":["Ahmet","Mehmet","Atil","Burak","Zeynep","Fatma"],
             "Maaslar":[100,200,150,200,300,400]
            }
```

```python
maasDataFrame=pd.DataFrame(maasSozlugu)
```

```python
maasDataFrame
```

```python
grupObjesi=maasDataFrame.groupby("Departman")
```

```python
grupObjesi.count()
```

```python
grupObjesi.mean()
```

```python
grupObjesi.min()
```

```python
grupObjesi.describe()
```
